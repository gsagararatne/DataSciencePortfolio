#### Importing sleep_score

Sleep efficiency and sleep score are 2 different metrics used by fitbit. Sleep efficiency was available through the API that I used, but the sleep score had to be taken directly through the fitbit website. The sleep score metric is what is displayed in the fitbit app. Personally helps me quantify my sleep and in fact, I seem to feel sick and groggy when I wake up from a sleep score below 80. That being said, it happens to be calculated through the metrics Time Asleep, Deep and REM, and Restoration.

Instructions on how to attain your own personal sleep score can be found through fitbit's [help page](https://help.fitbit.com/articles/en_US/Help_article/1133.htm?Highlight=export).

```python
sleep_score_df = pd.read_csv("sleep_score.csv")

# Check
sleep_score_df.head()
```

**Sleep score** is stored under the column `overall_score`. When extracting the csv file from fitbit, it gives me the entire history. That being said, I will need to filter out the certain rows of data that align with my timeline.

My method in mind would be to use the column `timestamp` as my filter.

```python
# Check timestamp data type
sleep_score_df["timestamp"].dtype
```

Since the data type of `timestamp` is an object. Will need to convert it into datetime.

```python
# convert to datetime
sleep_score_df["timestamp"] = pd.to_datetime(sleep_score_df["timestamp"])

print(sleep_score_df["timestamp"].dtype)

# Check format
sleep_score_df.head(1)
```

```python
# extract date only
sleep_score_df["timestamp"] = sleep_score_df["timestamp"].dt.date

# Check format
print(sleep_score_df["timestamp"].dtype)

sleep_score_df.head(1)
```

```python
# converting back to datetime
sleep_score_df["timestamp"] = pd.to_datetime(sleep_score_df["timestamp"])

# Check
print(sleep_score_df["timestamp"].dtype)


sleep_score_df.head()
```

`timestamp` finally only consists of date and is in datetime format. 

```python
# filtering rows

# criteria
start_date = "2021-02-25"
end_date = "2021-11-12"

mask = (sleep_score_df['timestamp'] >= start_date) & (sleep_score_df['timestamp'] <= end_date)

# applying mask
SleepScoreDf = sleep_score_df.loc[mask]

print(SleepScoreDf.shape)
SleepScoreDf.head()
```

One thing I've noticed is that the extracted rows only consists of 241 instead of 261 rows of data. Not sure as to why, but I'll take what I can get. 

Apart from that, the was the dataframe is organised in descending order in terms of timestamp. Will need to reverse the order given that it's usually the other way around. 

As of right now, the only 2 columns that interest me are `timestamp` and `overall_score`.

```python
# extracting 2 columns
SleepScoreDf = SleepScoreDf[["timestamp", "overall_score"]]

# reverse row order/index
SleepScoreDf_reverse = SleepScoreDf.reindex(index=SleepScoreDf.index[::-1])

# Reset index
SleepScoreDf_reverse.reset_index(drop=True, inplace=True)

# Check
SleepScoreDf_reverse
```

### Merging into one dataframe `fitbitdf`

Since the DataFrame **SleepScoreDf_reverse** has 241 rows of data compared to all the others (consisting of 261), I plan to concatenate all the DataFrames with 261 rows of data and then merge it with **SleepScoreDf_reverse**. Mainly because I plan to drop the incomplete rows of data.

```python
# concat dfs
fitbitdf = pd.concat([act_df, restingHR_df, sleep_df, steps_df], axis=1)

# dates index
dates = pd.date_range('2021-02-25', '2021-11-12')
fitbitdf.index = dates

print(fitbitdf.shape)

# Check
fitbitdf.head()
```

```python
fitbitdf.tail()
```

The dataframes with 261 rows of data have successfully concatenated. I've noticed that the sleep related data in 2021-11-12 are filled with Nans, and that's because of the way the data was extracted.

```python
# Note: cell only runs once

# index = timestamp
fitbitdf["timestamp"] = fitbitdf.index

# merge
fitbitdf = pd.merge(fitbitdf,SleepScoreDf_reverse, on="timestamp")

# reset index to timestamp
fitbitdf.index = fitbitdf["timestamp"]
```

```python
# check
print(fitbitdf.shape)
fitbitdf.head()
```

Successfully created the `fitbitdf` dataframe. Its index is the date of extraction. Before I decide to explore and/or alter this dataframe further, I'd create a column called `day_name` and `day_of_week` to distinguish the day name and the day of the week respectively.

### Feature engineering
#### Creating columns `day_name` and `day_of_week`

```python
# day of the week
fitbitdf["day_of_week"] = fitbitdf.index.weekday

# creating day name dictionary
days = {0:"Monday", 1:"Tuesday", 2:"Wednesday", 3:"Thursday", 4:"Friday", 5:"Saturday", 6:"Sunday"}

# day name
fitbitdf["day_name"] = fitbitdf["day_of_week"].apply(lambda dayOfWeek: days[dayOfWeek])

# Check
fitbitdf.head()
```

`day_name` and `day_of_week` have been successfully added. I would like to add a column called `time_period` which will indicate which timeline the date falls in. This columns is meant for filter purposes when I use *seaborn* to visualise my data. That being said, it'll also be split into binary columns. That way I can analyse my stats without heavily relying on the datetime index and potentially have more features to run ML models with.

The timeline which I'd like to implement are down below:
- `uni`: the period in which I was still in uni
    - 2021-02-25 : 2021-04-21
- `post_uni_pre_bootcamp` : the period after graduating uni
    - 2021-04-22 :  2021-06-06
- `bootcamp`: the period I was doing my data science bootcamp
    - 2021-06-07 : 2021-08-27
- `post_bootcamp`: the period I finished my bootcamp
    - 2021-08-28 : 2021-11-12
- `powerlifting`: the period I started powerlifting again
    - 2021-09-06 : 2021-11-12

```python
# uni criteria
uni_cri = fitbitdf.index <= "2021-04-21"

# post uni pre bootcamp criteria
post_uni_pre_boot_cri = ((fitbitdf.index > "2021-04-21") & (fitbitdf.index < "2021-06-07"))

# bootcamp criteria
bootcamp_cri = ((fitbitdf.index >= "2021-06-07") & (fitbitdf.index <= "2021-08-27"))

# post bootcamp criteria
post_boot_cri = fitbitdf.index > "2021-08-27"

#started powerlifting criteria
started_powerlifting_cri = fitbitdf.index >= "2021-09-06"
```

#### Creating `time_period`

```python
# filling up time_period
fitbitdf["time_period"] = "blah"
fitbitdf.loc[uni_cri,"time_period"] = "uni"
fitbitdf.loc[post_uni_pre_boot_cri,"time_period"] = "post_uni_pre_boot"
fitbitdf.loc[bootcamp_cri,"time_period"] = "bootcamp"
fitbitdf.loc[post_boot_cri,"time_period"] = "post_bootcamp"
```

### Question 3: sleep deficit 
I would like to know if I tend to sleep longer once I build a sleep deficit (not probably sleep) the night before. This is because I remember getting 3 hours of sleep during my finals in uni (for 3-4 consecutive nights) and then crashing for about 8-9 hours.

Will create a column called `yes_hours_asleep` and `hours_diff` to capture the number of hours slept the previous day and the sleep deficit built respectively.

Since amount of sleep is normally distributed, I can run a linear regression model to find a pattern.

```python
# offsetting by 1 - yesterda's sleep
fitbitdf["yes_hours_asleep"] = fitbitdf.shift(1).hours_asleep

# sleep deficit
fitbitdf["hours_diff"] = fitbitdf.hours_asleep - fitbitdf.yes_hours_asleep

fitbitdf.head()
```

```python
# linear regression model
sleep_deficit_linear_model = smf.ols(formula = 'hours_diff ~ yes_hours_asleep', data = fitbitdf).fit()
sleep_deficit_linear_model.summary().tables[0]
```

```python
# Plot how much more I slept each night vs. amount slept night before
sns.regplot(x = fitbitdf.yes_hours_asleep, y = fitbitdf.hours_diff, color = "mediumseagreen")
plt.title("Linear regression: sleep deficit")
plt.ylabel("Difference in Hours")
plt.xlabel("Hours of Sleep")
plt.axvline(7,c="black")
plt.axhline(0,c="black")
plt.show()
```

```python
plt.figure()
plt.scatter(sleep_deficit_linear_model.fittedvalues, sleep_deficit_linear_model.resid, color = "mediumseagreen")
plt.xlabel("Fitted values")
plt.ylabel("Residuals")
plt.title("Checking for constant cariance of residuals")
plt.show()
```

```python
plt.figure()
plt.hist(sleep_deficit_linear_model.resid, color = "mediumseagreen")
plt.title("Residuals histogram")
plt.ylabel("Frequency")
plt.show()
```

#### Question 3 answer:

The linear regression table shows me a negative coefficient followed by a p-value less than 0.05 (with an R squared value of 0.55). Meaning that I tend to sleep more the following day when I don't get a good amount of sleep the previous night.

On top of that, the model residuals are Homoscedastic and its histogram shows a normal distribution. Which is good! Happy days!

Now my sweet spot would be 7 hours of sleep a day.

```python
y2 = fitbitdf["good_sleep"]
X2 = fitbitdf[["sleep_efficiency", "hours_asleep", "wake_sleep"]]
```

```python
X_withconstant2 = sm.add_constant(X2)

# 1. Instantiate model
mylogreg= sm.Logit(y2,X_withconstant2)

# 2. Fit model
mylogreg_results = mylogreg.fit()

# Look at summary
mylogreg_results.summary()
```

```python
beta2 = mylogreg_results.params[2]
beta3 = mylogreg_results.params[3]

print(np.exp(beta2), np.exp(beta3))
```

For every additional hour of sleep, the odds of `good_sleep` increases by a likelihood of 3.8

For every additional hour of interruption (`wake_sleep`), the odds of `good_sleep` decreases by a likelihood of 43.8
(a more intuitive way of understanding this would be: For every additional hour of interruption (`wake_sleep`), the odds of `good_sleep` decreases by 97.7%)

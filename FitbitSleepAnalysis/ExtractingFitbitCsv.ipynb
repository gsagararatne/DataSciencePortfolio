##### Extracting resting heart rate data

```python
dates = pd.date_range('2021-02-25', '2021-11-12')

restingHR = []
for date in dates:
    fname = 'daily_HR' + date.strftime('%Y-%m-%d') + '.json'
    with open(fname) as f:
        restingHR_data = json.loads(f.read())
        restingHRdata = pd.read_json(restingHR_data, typ='series')
        restingHR.append(restingHRdata["activities-heart"][0]["value"]["restingHeartRate"]) # the zero indexes the first value
```

##### Extracting sleep data

```python
# Lists to store respective data
sleep_efficiency = []
deep_sleep = []
light_sleep = []
rem_sleep = []
wake_sleep = []
total_mins_asleep = []
total_time_in_bed = []

dates = pd.date_range('2021-02-25', '2021-11-12')

for date in dates:
    fname = 'sleep'+date.strftime('%Y-%m-%d')+'.json'
    with open(fname) as f:
        date_data_sleep = json.loads(f.read())
        datasleep = pd.read_json(date_data_sleep, typ='series')
        
        # sleep efficiency
        try:
            sleep_efficiency.append(datasleep["sleep"][0]["efficiency"])
        except:
            sleep_efficiency.append(np.nan)
        
        # deep sleep
        try:
            deep_sleep.append(datasleep["summary"]["stages"]["deep"])
        except:
            deep_sleep.append(np.nan)
           
            
        # light sleep
        try:
            light_sleep.append(datasleep["summary"]["stages"]["light"])
        except:
            light_sleep.append(np.nan)
            
            
        # rem sleep
        try:
            rem_sleep.append(datasleep["summary"]["stages"]["rem"])
        except:
            rem_sleep.append(np.nan)           
            
        # wake sleep
        try:
            wake_sleep.append(datasleep["summary"]["stages"]["wake"])
        except:
            wake_sleep.append(np.nan)                   

        # mins asleep    
        try:
            total_mins_asleep.append(datasleep['summary']['totalMinutesAsleep'])
        except:
            total_mins_asleep.append(np.nan)            
            
        # time in bed  
        try:
            total_time_in_bed.append(datasleep['summary']['totalTimeInBed'])
        except:
            total_time_in_bed.append(np.nan)
```

```python
# check for nan values
print("number of empty values for sleep efficiency: ", np.isnan(sleep_efficiency).sum())
print("number of empty values for deep sleep: ", np.isnan(deep_sleep).sum())
print("number of empty values for light sleep: ", np.isnan(light_sleep).sum())
print("number of empty values for rem sleep: ", np.isnan(rem_sleep).sum())
print("number of empty values for wake sleep: ", np.isnan(wake_sleep).sum())
print("number of empty values for total mins asleep: ", np.isnan(total_mins_asleep).sum())
print("number of empty values for total time in bed: ", np.isnan(total_time_in_bed).sum())
```

```python
print("list length of sleep efficiency: ", len(sleep_efficiency))

print("list length of deep sleep: ", len(deep_sleep))

print("list length of light sleep: ", len(light_sleep))

print("list length of rem sleep: ", len(rem_sleep))

print("list length of wake sleep: ", len(wake_sleep))

print("list length of total mins asleep: ", len(total_mins_asleep))

print("list length of total time in bed: ", len(total_time_in_bed))
```

My sleep data consists of 261 rows of data. I had to implement **try** and **except** statements in my extracting loop because `stages` was absent under `summary` in the `datasleep` dictionary for 7 json files. This is because I forgot to wear my fitbit to sleep. Hence, had to replace said missing values with **Nans**.

```python
# Creating the sleepdf dataframe
sleepdf = pd.DataFrame([sleep_efficiency, deep_sleep, light_sleep, rem_sleep, wake_sleep, total_mins_asleep, total_time_in_bed]) 
sleepdf = sleepdf.transpose()
sleepdf.columns = ["sleep_efficiency", "deep_sleep", "light_sleep", "rem_sleep", "wake_sleep", "total_mins_asleep", "total_time_in_bed"]
sleepdf = sleepdf.set_index(dates)
print(sleepdf.shape)

# Check
sleepdf.head()
```

##### Extacting activity data

```python
# Lists to store respective data
calories_burnt = []
distance_km = []


# dates = pd.date_range('2021-02-25', '2021-02-26')
dates = pd.date_range('2021-02-25', '2021-11-12')

for date in dates:
    fname = 'act'+date.strftime('%Y-%m-%d')+'.json'
    with open(fname) as f:
        date_data_act = json.loads(f.read())
        dataact = pd.read_json(date_data_act, typ='series')
            
        calories_burnt.append(dataact["summary"]["caloriesOut"])   
        distance_km.append(dataact["summary"]["distances"][0]["distance"])
```

`calories_burnt` and `distance_km` has been successfully filled with calories burnt and distance walked in a day (km) respectively. They both consist of 261 data points with no Nan values.

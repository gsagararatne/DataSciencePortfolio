### Missing data EDA
The purpose of this section isn't to completely clean data but rather explore and make decisions accordingly. I will most probably delete rows out of convenience mainly because I don't want to fill in information blindly. Example: station ids can potentially change over the years.

```python
BikeShareToronto_df.isna().sum()
```

As of right now, dropping `from_station_id` and `to_station_id` would be very convenient. It's more intuitive to know the name of the stop than its digital counterpart. However, I'll also create a block of code to check if ids of locations were changed over time.

```python
# names of start stations that have missing ids
BikeShareToronto_df[BikeShareToronto_df["from_station_id"].isna()]["from_station_name"].value_counts()
```

```python
# names of end stations that have missing ids
BikeShareToronto_df[BikeShareToronto_df["to_station_id"].isna()]["to_station_name"].value_counts()
```

Union station seems to have the most missing values.

```python
# number of end IDs union station has (had)
BikeShareToronto_df[BikeShareToronto_df["to_station_name"]=="Union Station"]["to_station_id"].value_counts()
```

Union station in this case seems to have the ID 7033. In this case, I could fill the missing Nans for Union station with 7033.

### Changing data types

```python
BikeShareToronto_df.info()
```

```python
BikeShareToronto_df["trip_start_time"] = pd.to_datetime(BikeShareToronto_df["trip_start_time"], errors='coerce')
```

```python
BikeShareToronto_df["trip_stop_time"] = pd.to_datetime(BikeShareToronto_df["trip_stop_time"], errors='coerce')
```

```python
BikeShareToronto_df["start_date"] = BikeShareToronto_df["trip_start_time"].dt.strftime('%d/%m/%Y')
BikeShareToronto_df["start_time"] = BikeShareToronto_df["trip_start_time"].dt.strftime('%H:%M:%S')
BikeShareToronto_df["end_date"] = BikeShareToronto_df["trip_stop_time"].dt.strftime('%d/%m/%Y')
BikeShareToronto_df["end_time"] = BikeShareToronto_df["trip_stop_time"].dt.strftime('%H:%M:%S')
```
